```python
import pandas as pd
import os 
import requests
import json
```

```python
catalog=pd.read_json("catalog.json")
catalog
```

# CREATING REQUIRED DIRECTORIES


```python
os.mkdir("../images")
for  cat in catalog["category"].unique():
      os.mkdir(f"../images/{cat}")
```

# DOWNLOADING IMAGES AND STORING IN CORRESPONDING DIRECTORIES

```python
# downloading_image
def download_image(url, filename):
    response = requests.get(url, stream=True)
    response.raise_for_status()

    with open(filename, 'wb') as f:
        for chunk in response.iter_content(1024):
            f.write(chunk)
```

```python
download_folder="../images"
for index,row in catalog.iterrows():
    filepath=f"{download_folder}/{row.category}/{row.product_name}.jpg"
    
    download_image(row.image_url,filepath)
    row.image_url=filepath
```

```python
catalog["image_url"]=download_folder+"/"+catalog["category"]+"/"+catalog["product_name"]+".jpg"
```
