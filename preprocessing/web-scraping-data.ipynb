```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def scrape_data(category:str,url:str,catalog:list):
        soup=BeautifulSoup(requests.get(url).content,'html.parser')
        product_containers = soup.find_all('div', class_='product-thumb size-option')
        for container in product_containers:
      
                image_tag = container.find('img', class_='swap-image lload img-responsive')
                image_url = image_tag['data-src']

                price_tag=container.find("span",class_="price-new")
                product_name=container.find("h4",class_="product-name")
                product_info_link=container.find("a",class_="single-product")["href"]
                product_info_request=requests.get(product_info_link)

                product_info_html=BeautifulSoup(product_info_request.content,"html.parser")

                catalog["description"].append(product_info_html.find_all("p")[2].text.strip())
                catalog["product_name"].append(product_name.text.strip())  
                catalog["image_url"].append(image_url)
                catalog["category"].append(category)

                if price_tag is  not None :
                        catalog["price"].append(price_tag.text.strip())
                else:
                        price_tag=container.find("p",class_="price")
                        catalog["price"].append(price_tag.text.strip())  
    
        return catalog 

```

```python
catalog={"product_name":[],"price":[],"category":[],"description":[],"image_url":[]}
mens_shirts_url="https://www.jackjones.in/shirts-jj"
mens_tshirts_url="https://www.jackjones.in/t-shirts-jj"
mens_pants_url="https://www.jackjones.in/jeans-jj"
mens_jackets_url="https://www.jackjones.in/jeans-jj"
mens_footwear_url="https://www.jackjones.in/jeans-jj"
mens_caps_url="https://www.jackjones.in/accessories-jj/caps-accessories-jj"
mens_backpacks_url="https://www.jackjones.in/accessories-jj/mens-backpacks"
mens_wallets_url="https://www.jackjones.in/accessories-jj/mens-wallets"
mens_sunglasses_url="https://www.jackjones.in/accessories-jj/jj-sunglasses"
catalog=scrape_data(category="shirts",url=mens_shirts_url,catalog=catalog)
catalog=scrape_data(category="tshirts",url=mens_tshirts_url,catalog=catalog)
catalog=scrape_data(category="pants",url=mens_pants_url,catalog=catalog)
catalog=scrape_data(category="jackets",url=mens_jackets_url,catalog=catalog)
catalog=scrape_data(category="footwear",url=mens_footwear_url,catalog=catalog)
catalog=scrape_data(category="caps",url=mens_caps_url,catalog=catalog)
catalog=scrape_data(category="backpacks",url=mens_backpacks_url,catalog=catalog)
catalog=scrape_data(category="wallets",url=mens_wallets_url,catalog=catalog)
catalog=scrape_data(category="sunglasses",url=mens_sunglasses_url,catalog=catalog)

```

```python
df=pd.DataFrame(catalog)
```
